```python
from fastai.vision.all import *
import matplotlib.pyplot as plt

# Root path to dataset
root_path = Path('breast_cancer_data')

# Collect image paths recursively for class 0 and class 1
benign_images = get_image_files(root_path).filter(lambda f: '/0/' in f.as_posix())[:5]
malignant_images = get_image_files(root_path).filter(lambda f: '/1/' in f.as_posix())[:5]

# Helper function to display images
def plot_images(img_paths, title, nrows=1):
    n = len(img_paths)
    plt.figure(figsize=(n * 3, 3))
    for i, img_path in enumerate(img_paths):
        img = PILImage.create(img_path)
        plt.subplot(nrows, n, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"{title} {i+1}")
    plt.show()

# Plot benign and malignant images
plot_images(benign_images, title="Benign")
plot_images(malignant_images, title="Malignant")


```
